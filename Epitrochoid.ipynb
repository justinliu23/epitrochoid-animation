```python
from numpy import sin, cos, linspace, pi
import pylab

# curve parameters
R = 5
r = 1
d = 2

t = linspace(0, 2 * pi, 300)

# Epitrochoid parametric equations
x = (R - r)*cos(t) - d * cos((R + r) * t / r)
y = (R - r) * sin(t) - d * sin((R + r) * t / r)

# plot
pylab.plot(x, y, 'r')
pylab.axis('equal')
pylab.show()
```
